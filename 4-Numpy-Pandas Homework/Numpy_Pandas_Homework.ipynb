```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data = "PyCitySchools.csv"
student_data = "PyCityStudents.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data)
student_data = pd.read_csv(student_data)
student_data.head()
```

```python
# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head(5)
```

```python
#*******************************************************************************OVERALL school_data_complete.info()
```

```python
grouped_school_data = school_data_complete.groupby(['school_name'])
total_budget = 0

total_schools = len(school_data_complete["school_name"].unique())
total_students = (school_data_complete["student_name"].count())

total_budget = (grouped_school_data["budget"].first() + total_budget)
total_budget1 = total_budget.sum()

avg_math = school_data_complete["math_score"].mean()
avg_reading = school_data_complete["reading_score"].mean()

student_data['pass_math'] = student_data['math_score'].apply(lambda x: x >= 70)
pass_math_count = student_data['pass_math'].value_counts()[1] 
per_math_pass=(float(pass_math_count) / total_students)*100 

student_data['pass_reading'] = student_data['reading_score'].apply(lambda x: x >= 70)
pass_reading_count = student_data['pass_reading'].value_counts()[1] 
per_reading_pass=(float(pass_reading_count) / total_students)*100 

overall_pass_rate = ((avg_math + avg_reading)/2)
```

```python

summary_table = pd.DataFrame({"Total Schools": total_schools,
                              "Total Students": [total_students],
                              "Total Budget": [total_budget1],  
                              "Average Math Score": [avg_math],
                              "Average Reading Score":[avg_reading],
                              "% Passing Math": [per_math_pass],
                              "% Passing Reading": [per_reading_pass],
                              "% Overall Passing Rate":[overall_pass_rate]})
                                     
summary_table1 = pd.DataFrame(summary_table)
summary_table1['Total Budget'] = summary_table1['Total Budget'].map('${:,.2f}'.format) 
summary_table1['% Passing Math'] = summary_table1['% Passing Math'].map('%{:,.2f}'.format)
summary_table1['% Passing Reading'] = summary_table1['% Passing Reading'].map('%{:,.2f}'.format) 
summary_table1['% Overall Passing Rate'] = summary_table1['% Overall Passing Rate'].map('%{:,.2f}'.format) 


summary_table1
```

```python
#*********************************************************************************************PER SCHOOL FROM HIGHEST

grouped_school_data = school_data_complete.groupby(['school_name'])

#---------------------------------------------------------------------------

#school_counts = grouped_school_data["school_name"].value_counts()
school_counts = grouped_school_data["size"].first()

school_type = grouped_school_data["type"].first()
school_budgets = grouped_school_data["budget"].first()
per_student_budget = school_budgets/school_counts
avg_math_scores = grouped_school_data["math_score"].mean()
avg_reading_scores = grouped_school_data["reading_score"].mean()

filtered_math_pass = school_data_complete[school_data_complete["math_score"]>=70]
high_pass_math_count = filtered_math_pass.groupby("school_name").count()
high_test_math = (high_pass_math_count["math_score"]/school_counts)*100

filtered_reading_pass = school_data_complete[school_data_complete["reading_score"]>=70]
high_pass_reading_count = filtered_reading_pass.groupby("school_name").count()
high_test_reading = (high_pass_reading_count["reading_score"]/school_counts)*100

high_overall_pass_rate = (high_test_reading + high_test_math)/2
```

```python

schools_summary_table = pd.DataFrame({"School Type": school_type,
                                    "Total Students": school_counts,
                                     "Total School Budget": school_budgets,
                                     "Per Student Budget": per_student_budget,
                                     "Average Math Scores": avg_math_scores,
                                     "Average Reading Scores": avg_reading_scores,
                                     "% Passing Math": high_test_math,
                                     "% Passing Reading": high_test_reading,
                                     "% Overall Passing Rate": high_overall_pass_rate})

schools_summary_table.head()
high_df = pd.DataFrame(schools_summary_table)

high_df1 = high_df.sort_values(by=['% Overall Passing Rate'], ascending=False)

high_df1['Total School Budget'] = high_df1['Total School Budget'].map('${:,.2f}'.format) 
high_df1['% Passing Math'] = high_df1['% Passing Math'].map('%{:,.2f}'.format)
high_df1['% Passing Reading'] = high_df1['% Passing Reading'].map('%{:,.2f}'.format) 
high_df1['% Overall Passing Rate'] = high_df1['% Overall Passing Rate'].map('%{:,.2f}'.format) 

high_df1.head(5)
```

```python
#************************************************************************************************PER SCHOOL FROM LOWEST
schools_summary_table = pd.DataFrame({"School Type": school_type,
                                    "Total Students": school_counts,
                                     "Total School Budget": school_budgets,
                                     "Per Student Budget": per_student_budget,
                                     "Average Math Scores": avg_math_scores,
                                     "Average Reading Scores": avg_reading_scores,
                                     "% Passing Math": high_test_math,
                                     "% Passing Reading": high_test_reading,
                                     "% Overall Passing Rate": high_overall_pass_rate})

schools_summary_table
low_df = pd.DataFrame(schools_summary_table)
low_df1 = low_df.sort_values(by=['% Overall Passing Rate'], ascending=True)

low_df1['Total School Budget'] = low_df1['Total School Budget'].map('${:,.2f}'.format) 
low_df1['Per Student Budget'] = low_df1['Per Student Budget'].map('${:,.2f}'.format) 
low_df1['% Passing Math'] = low_df1['% Passing Math'].map('%{:,.2f}'.format)
low_df1['% Passing Reading'] = low_df1['% Passing Reading'].map('%{:,.2f}'.format) 
low_df1['% Overall Passing Rate'] = low_df1['% Overall Passing Rate'].map('%{:,.2f}'.format) 

low_df1.head(5)
```

```python
#*************************************************************************************************PERFORMANCE P/STUDENT

# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
spending_per_student = school_budgets / school_counts

#^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^
per_student_summary_table = pd.DataFrame({"Average Math Scores": avg_math_scores,
                                     "Average Reading Scores": avg_reading_scores,
                                     "% Passing Math": high_test_math,
                                     "% Passing Reading": high_test_reading,
                                     "% Overall Passing Rate": high_overall_pass_rate,
                                     "per/student spend performance": spending_per_student})

df = pd.DataFrame(per_student_summary_table)
df["Spending Ranges Per Student"]= pd.cut(df["per/student spend performance"], spending_bins, labels = group_names)
```

```python

df1 = df.sort_values(by=['Spending Ranges Per Student'])
df1.reset_index()

df_grouped = df1.groupby(["Spending Ranges Per Student"])
#df_grouped.mean()
grouped_spending_per_student_mean = df_grouped.mean()
#grouped_spending_per_student_mean.loc[0:4,"Average Math Scores":"% Overall Passing Rate"]
grouped_spending_per_student_mean.drop("per/student spend performance", axis =1)
```
